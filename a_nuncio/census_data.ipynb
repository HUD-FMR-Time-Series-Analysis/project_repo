The data sets have a x-axis label but the label was read in as a column. The column: Label (Grouping), has been set as the index.

```python
# set the index of each data frame to be column: 'Label (Grouping)'
df_10 = df_10.set_index('Label (Grouping)')
df_11 = df_11.set_index('Label (Grouping)')
df_12 = df_12.set_index('Label (Grouping)')
df_13 = df_13.set_index('Label (Grouping)')
df_14 = df_14.set_index('Label (Grouping)')
df_15 = df_15.set_index('Label (Grouping)')
df_16 = df_16.set_index('Label (Grouping)')
df_17 = df_17.set_index('Label (Grouping)')
df_18 = df_18.set_index('Label (Grouping)')
df_19 = df_19.set_index('Label (Grouping)')
df_20 = df_20.set_index('Label (Grouping)')
df_21 = df_21.set_index('Label (Grouping)')
```

Several changes need to be made to column names for readablity and simplicity. I will be formating column names to be lowercase, snakecase, abreviated, and I will remove any unecessary charachters.

```python
# format column names to be lowercase
df_10.columns = df_10.columns.str.lower()
# format column names to be snakecase
df_10.columns = df_10.columns.str.replace(' ', '_')
# replace !! with snakecase
df_10.columns = df_10.columns.str.replace('!!', '_')
# remove commas
df_10.columns = df_10.columns.str.replace(',', '')
# shorten texas to tx
df_10.columns = df_10.columns.str.replace('texas', 'tx')
# shorten san antonio to sa
df_10.columns = df_10.columns.str.replace('san_antonio', 'sa')
# shorten margin of error to moe
df_10.columns = df_10.columns.str.replace('margin_of_error', 'moe')
# shorten estimate to est
df_10.columns = df_10.columns.str.replace('estimate', 'est')
```

There are several labels used to label subgroups in the dataframe but do not contain information its self. I will drop these below.

```python
# drop EMPLOYMENT STATUS
df_10.drop(labels='EMPLOYMENT STATUS', inplace=True)
# drop COMMUTING TO WORK
df_10.drop(labels='COMMUTING TO WORK', inplace=True)
# drop OCCUPATION
df_10.drop(labels='OCCUPATION', inplace=True)
# drop INDUSTRY
df_10.drop(labels='INDUSTRY', inplace=True)
# drop CLASS OF WORKER
df_10.drop(labels='CLASS OF WORKER', inplace=True)
# drop INCOME AND BENEFITS (IN 2010 INFLATION-ADJUSTED DOLLARS)
df_10.drop(labels='INCOME AND BENEFITS (IN 2010 INFLATION-ADJUSTED DOLLARS)', inplace=True)
# drop HEALTH INSURANCE COVERAGE
df_10.drop(labels='HEALTH INSURANCE COVERAGE', inplace=True)
# drop PERCENTAGE OF FAMILIES AND PEOPLE WHOSE INCOME IN THE PAST 12 MONTHS IS BELOW THE POVERTY LEVEL
df_10.drop(labels='PERCENTAGE OF FAMILIES AND PEOPLE WHOSE INCOME IN THE PAST 12 MONTHS IS BELOW THE POVERTY LEVEL', inplace=True)
```
