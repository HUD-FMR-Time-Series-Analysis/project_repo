Several changes need to be made to column names for readablity and simplicity. I will be formating column names to be lowercase, snakecase, abreviated, and I will remove any unecessary charachters.

```python
dataframes = [df_10, df_11, df_12, df_13, df_14, 
              df_15, df_16, df_17, df_18, df_19, 
              df_20, df_21]
# for loop
for df in dataframes:
    # format column names to be lowercase
    df.columns = df.columns.str.lower()
    # format column names to be snakecase
    df.columns = df.columns.str.replace(' ', '_')
    # replace !! with snakecase
    df.columns = df.columns.str.replace('!!', '_')
    # remove commas
    df.columns = df.columns.str.replace(',', '')
    # shorten texas to tx
    df.columns = df.columns.str.replace('texas', 'tx')
    # shorten san antonio to sa
    df.columns = df.columns.str.replace('san_antonio', 'sa')
    # shorten margin of error to moe
    df.columns = df.columns.str.replace('margin_of_error', 'moe')
    # shorten estimate to est
    df.columns = df.columns.str.replace('estimate', 'est')
    # remove (
    df.columns = df.columns.str.replace('(', '')
    # remove )
    df.columns = df.columns.str.replace(')', '')
```
