```python
# list of columns from df1 that I will drop
cols_to_drop = ['County Number', 'ESC Region Served', 'ESC Region PEIMS', 'ESC Region Geographic', 
                'NCES District ID', 'District Street Address', 'District City', 'District State',
                'District Zip', 'District Phone', 'District Fax', 'District Email Address', 
                'District Web Page Address', 'District Superintendent', 'Residential Facility', 
                'NCES School ID', 'School Phone', 'School Fax', 'School Email Address', 
                'School Web Page Address', 'School Principal', 'School Status', 'School Status Date',
                'Update Date', 'School Street Address', 'School City', 'School State', 
                'District Enrollment as of Oct 2022', 'Grade Range']
# drop list of columns from df
df1 = df1.drop(columns=cols_to_drop)
```

```python
# format column names to be lowercase and snakecase
df1.columns = df1.columns.str.lower().str.replace(' ', '_')

# make df lowercase and snake case
df1 = df1.applymap(lambda x: x.lower() if isinstance(x, str) else x)

# remove single quote from df values
df1 = df1.replace("'", '', regex=True)

# remove -0000 from zip codes and insert the modified zips as a new column
df1['zipcode'] = df1['school_zip'].str.split('-').str[0]

# drop 'school_zip' from df1
df1 = df1.drop(columns=['school_zip'])

# change numeric columns to ints
to_int = ['district_number', 'school_number', 'zipcode']
for col in to_int:
    df1[col] = df1[col].astype(int)

# rename columns
df1 = df1.rename(columns={'county_name': 'county',
                     'district_number': 'district_num',
                     'district_name': 'dist_name',
                     'district_type': 'dist_type',
                     'school_number': 'school_num',
                     'instruction_type': 'instruct_type',
                     'school_enrollment_as_of_oct_2022': 'school_enrollment'})
```

```python
# isolate the data frame to only show zip codes in my zip_codes_int list
df1_iced = df1[df1['zipcode'].isin(zip_codes_int)]

# list of zip codes in df1
df1_zips = df1['zipcode'].tolist()
# list of unique zip codes in df1
df1_uz = df1['zipcode'].unique().tolist()

# a list of zip codes in df1_iced
df1_iced_zips = df1_iced['zipcode'].tolist()
# a list of unique zip codes in df1_iced
df1_iced_uz = df1_iced['zipcode'].unique().tolist()

# count of zip codes in zip_codes list
count_zips = len(zip_codes_int)
print(f'Count of zip codes in zip_codes list: \n{count_zips}')
print('-'*55)

# count of zip codes in df1_iced
count_df1_iced = len(df1_iced_zips)
print(f'Count of zip codes in df1_iced: \n{count_df1_iced}')
print('-'*55)

# count of unique zip codes in df1_iced
uz_count_df1_iced = len(df1_iced_uz)
print(f'Count of unique zip codes in df1_iced: \n{uz_count_df1_iced}')
print('-'*55)

# List of zip codes, from zip_codes list, not in df1_iced.
not_in_df1_iced = list(set(zip_codes_int) - set(df1_iced_uz))
# count of zip codes from zip_codes list not in df1_iced
count_not_in_df1_iced = len(not_in_df1_iced)
print(f'Count of zip codes, from zip_codes list, not in df1_iced: \n{count_not_in_df1_iced}')
print('-'*55)

# percent of zips from my list of zips I was able to secure school data for
print(f'Percent of zips from my list of zips I was able to secure school data for: \n{42/138*100}')
```

```python
# format column names to be lowercase, snakecase, and remove special characters
soar.columns = soar.columns.str.lower().str.replace('\n', '_').str.replace(' ', '_').str.replace('/', '_')

# make df lowercase and snake case
soar = soar.applymap(lambda x: x.lower() if isinstance(x, str) else x)

# remove % from df values
soar = soar.replace("%", '', regex=True)

# make soar['campus_number'] dtype int
soar['campus_number'] = soar['campus_number'].fillna(0).astype(int)

# list of columns with object dtypes to change to float dtypes
to_float = ['%_economically_disadvantaged', '%_eb_el_students',]
# change dtype to float
for col in to_float:
    soar[col] = soar[col].astype(float)
    
# rename
soar = soar.rename(columns= {'campus_number': 'school_num',
                             '%_economically_disadvantaged': 'pcent_econ_disadv', 
                             '%_eb_el_students': 'pcent_esl', 
                             'overall_rating': 'rating'})
```
